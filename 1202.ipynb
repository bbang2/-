```python
import numpy as np
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs

from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
```

```python
url = 'https://sports.news.naver.com/wfootball/record/index.nhn'
driver = wb.Chrome()
driver.get(url)
soup = bs(driver.page_source, 'lxml')
```

```python
x=soup.select('td')
x
```

```python
data_list = []
for i in x:
    data_list.append(i.text.strip())
```

```python
data_list
```

```python
data=np.array(data_list).reshape(-1,10)
```

### 1.  del안하고 데이터 변환하기

```python
data[0][0]
```

```python
for i in data:
    print(len(i[0]))
    # 2보다 큰 값은 바꾸기
```

```python
cnt=0
for i in data:
    if len(i[0])>2:
        i[0]= cnt+1
        
    cnt+=1
```
